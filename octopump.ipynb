<h1>Octopump</h1> <br> So many open source pumps currently exist, so why add one more to the pile? I did it because none are as barebones simple and yet still flexible as this one is. This particular system was designed to accomodate 4 independant pumps, but in theory, the 3D printed housing could be modified and expanded to hold more (the motor drivers can be stacked together and each will drive up to 4 DC non-stepping motors, two if they are steppers). The housing was designed to hold all 4 pumps, a B+ raspberry pi 3 and the required motor driver. In addition to this, a custom cooling lid is available to add onto the system. Materials not including wire, screws and tools is approximately <b>$130 dollars</b> as of Oct 5th 2020.  

<h1>Ideas that Drove Project</h1> <br> 
1. No <b> GUI </b>! This may seem counter intutive at first glance, but while the omission of this feature decreases the ease of use for the tasks that the GUI was generated for, it greatly eases access for non-advanced users who desire to use the pumps for tasks other than what the GUI was intended to do. <br>
2. Only use python code <br>
3. Minimum of 4 pumps <br>
4. Bare bones easy to use methods <br>
5. All contained, compact system for portability

<h2>Pulse Width Modulation (PWM)</h2>

This issue may stem from an oversight on my part, but even thought the Adafruit motor HAT claimed full PWM function, I was not able to achieve it. The idea behind PWM is an effort to circumvent the decreasing torque issue that DC motors have when their voltage is decreased. It slows them down, but then they can decrease their torque so much that it stalls out instead of just slowing down. PWM is a method where the motor is turned on at full power, but then shut off too. It does this frequently and creates an average that is slower, but when its on, its at full torque. I created the following function in an effort to create this ON-OFF scenario. <h1>Applications</h1>

Here are some example applications of the pump system.

1. Periodically refresh media in a microfluidic device or biochamber while imaging a sample within it. The pumps can be set to only activate between images so that the sample isnt disturbed during the imaging. 

2. Set normal media to flow into an imaging and then after 1 hr., change media for media with inhibitor add into it. 

3. Auto slide washer/bleaching station. We can make a chamber that we place slides into it and the pumps wash PBS over the slides and or add on a fluorescence bleaching solution as in CyCIF. 

<h1>Super Slow FlowRate Issue</h1>

The methods I decribed have an issue when using very slow flow rates. It appears that in each on-off cycle that the pump does, it seems to have to move a minimum amount. So if you do a fast frequency "pulse", it pumps way more fluid than expected. To combat this, we must make the pumps more granular and reduce the frequency. 5 hz seems to do a good job.
